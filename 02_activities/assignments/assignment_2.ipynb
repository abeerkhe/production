# Load the data

Place the files in the ../../05_src/data/fires/ directory and load the appropriate file. 

```python
%load_ext dotenv 
%dotenv 

import os 

import sys
sys.path.append(os.getenv('SRC_DIR'))

import dask.dataframe as dd 
import pandas as pd
import numpy as np

from glob import glob

from sklearn.pipeline import Pipeline 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler, OneHotEncoder, PowerTransformer
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Lasso, Ridge
from sklearn.feature_selection import VarianceThreshold
from sklearn.impute import SimpleImputer

import pickle

import shap
```

```python
# Load data
columns = [
    'coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain', 'area']
fires_dt = (pd.read_csv('../../05_src/data/fires/forestfires.csv', header = 0, names = columns))
fires_dt.info()
```

# Get X and Y

Create the features data frame and target data.

```python
#Create the target data, area, as Y: 
y = fires_dt['area']
print(f"Null values: {y.isnull().values.any()}")
```

```python
#Create the features data frame, everything excelt the area, as X: 
x = fires_dt.drop(columns = 'area')
print(f"Null values: {x.isnull().values.any()}")
```

# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


```python
y.hist()
print(f"Max value: {y.max()} \nMedian value: {y.median()} \nMean value: {y.mean()}")

print("A low median, substantially higher mean, and extremely high max value (relative to the median and mean) would indicate using \na scaling method not overly sensitive to outliers. By process of elimination, this is the robust scaling method.")
```

### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
#Identify numerical and categorical features, based on the variable descriptions 
num_feat = ['coord_x', 'coord_y','ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain']
cat_feat = ['month', 'day']

#Nothing to account for NaN values as the dataset does not have nulls, as checked above for x and y dataframes. 

num_transformer1 = Pipeline([
    ('remove_zero_variance', VarianceThreshold()),  
    ('scaler', RobustScaler()),
    ('imputer', SimpleImputer(strategy='median'))                   
    ])

cat_transformer = Pipeline([('onehot', OneHotEncoder(handle_unknown='ignore'))]) #Single cat_transformer as preproc 1 and 2 both use the same categorical one-hot encoding 

preproc1 = ColumnTransformer([('num', num_transformer1, num_feat), ('cat', cat_transformer, cat_feat)],remainder='passthrough')
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
x.hist(bins=x.shape[1], figsize=(10,7))

print(f"Given the skewness of data below and background from the FWI system (i.e., inputs to each of our feature variables), \nthe values I chose for preproc2 are 'rain' (rain amount), 'isi' (initial spread, from wind), and 'dmc' (duff moisture, from temp, relative humidity, rain).")
```

```python
#The same cat_feat lists apply. 
#The same cat_transformer applies. 

num_feat2 = ['rain', 'isi', 'dmc']
num_transformer2 = Pipeline([('variance_threshold', VarianceThreshold()),
    ('log_transform', PowerTransformer()),
    ('imputer', SimpleImputer(strategy='median'))
    ])

preproc2 = ColumnTransformer([('nonlin_num', num_transformer2, num_feat2), 
                            ('cat', cat_transformer, cat_feat)
                            ], remainder='passthrough')
```

## Model Pipeline


Create a [model pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html): 

+ Add a step labelled `preprocessing` and assign the Column Transformer from the previous section.
+ Add a step labelled `regressor` and assign a regression model to it. 

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
#Split x and y into testing and training data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Pipeline A = preproc1 + baseline
print(f"I opted for Lasso instead of Ridge: \nLasso helps reduce overfitting to the model which can be valuable given the amount of variables in the dataset, and \nRidge helps account for multicolinnearity in the model which I am less concerned about given the variety of variables here.")
pipeline_A = Pipeline([('preprocessing', preproc1), ('regressor', Lasso(random_state=42))])
```

```python
# Pipeline B = preproc2 + baseline
print(f"Here I opted for Ridge considering the limited amount of variables with potential colinearity of 'rain' and 'isi' with 'dmc', \ngiven the inputs to calculate the dmc described above.")
pipeline_B = Pipeline([('preprocessing', preproc2), ('regressor', Ridge(random_state=42))])
```

```python
# Pipeline C = preproc1 + advanced model
print(f"Leveraging GradientBoosting to minimize overfitting and ideally have a more accurate predictive model.")
pipeline_C = Pipeline([('preprocessing', preproc1), ('regressor', GradientBoostingRegressor(random_state=42))])
```

```python
# Pipeline D = preproc2 + advanced model
pipeline_D = Pipeline([('preprocessing', preproc2), ('regressor', GradientBoostingRegressor(random_state=42))])
    
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
#Pipeline A, providing different values of alpha for the regressor. 
param_grid_A = {
    'regressor__alpha': [0.01, 0.1, 1.0, 10.0, 100.0]
}

#GridSearch for Pipeline A, using negative root mean square errors for scoring. 
grid_A = GridSearchCV(
    pipeline_A, param_grid_A, scoring='neg_root_mean_squared_error'
)

#Fitting the model 
grid_A.fit(x_train, y_train)

#print(grid_A.best_params_)
grid_A.best_estimator_
```

```python
#Pipeline B
param_grid_B = {
    'regressor__alpha': [0.1, 1.0, 10.0, 100.0]
}

grid_B = GridSearchCV(
    pipeline_B, param_grid_B, scoring='neg_root_mean_squared_error'
)

grid_B.fit(x_train, y_train)

#print(grid_A.best_params_)
grid_B.best_estimator_
```

```python
#Pipeline C
param_grid_C = {
    'regressor__n_estimators': [10, 50, 100, 150, 200],
    'regressor__learning_rate': [0.01, 0.1, 0.25, 0.5]
}

grid_C = GridSearchCV(
    estimator = pipeline_C,
    param_grid=param_grid_C,
    scoring='neg_root_mean_squared_error'
)

grid_C.fit(x_train, y_train)
grid_C.best_estimator_
```

```python
#Pipeline D
param_grid_D = {
    'regressor__n_estimators': [10, 50, 100, 150, 200],
    'regressor__learning_rate': [0.01, 0.1, 0.25, 0.5]
}

grid_D = GridSearchCV(
    estimator=pipeline_D,
    param_grid=param_grid_D,
    scoring='neg_root_mean_squared_error'
)

grid_D.fit(x_train, y_train)
grid_D.best_estimator_
```

# Evaluate

+ Which model has the best performance?

```python
#Collect results from all the pipelines for their best scores (root mean squared error) and parameter settings. 
res = {
    'Pipeline': ['A', 'B', 'C', 'D'],
    'Best RMSE': [grid_A.best_score_, grid_B.best_score_, grid_C.best_score_, grid_D.best_score_], 
    'Best parameter': [grid_A.best_params_, grid_B.best_params_, grid_C.best_params_, grid_D.best_params_]
}

res = pd.DataFrame(res)
res = res.sort_values('Best RMSE')
display(res)

print(f"The model with the best performance is Pipeline {res.iloc[0]['Pipeline']} with the best parameters of {res.iloc[0]['Best parameter']} and a score of {res.iloc[0]['Best RMSE']}.")
```

# Export

+ Save the best performing model to a pickle file.

```python
#Saving the best model as a variable, which was Pipeline C 
best_model = grid_C.best_estimator_
model_file = 'best_fire_model.pickle'

#Save as a pickle file
with open(model_file, 'wb') as file:
    pickle.dump(best_model, file)
```
